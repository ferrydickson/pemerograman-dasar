```python
def hitung_efisiensi(input_energi, output_energi):
  """Menghitung efisiensi turbin berdasarkan input dan output energi.
  Parameters:
  input_energi (float): Energi input dalam satuan tertentu
  output_energi (float): Energi output dalam satuan tertentu
  Returns:
  float: Nilai efisiensi (Output / Input)"""
  efisiensi = output_energi / input_energi
  return efisiensi

# Bonus: Memanggil fungsi dengan input 500 dan output 400
input_energi = 500
output_energi = 400
hasil_efisiensi = hitung_efisiensi(input_energi, output_energi)

print(f"Input Energi: {input_energi}")
print(f"Output Energi: {output_energi}")
print(f"Efisiensi Turbin: {hasil_efisiensi:.2f} atau {hasil_efisiensi*100:.1f}%")
```

```python
# Variabel untuk tegangan dan arus
tegangan = 220  # dalam Volt
arus = 15       # dalam Ampere

# Ekspresi untuk menghitung daya listrik
daya = tegangan * arus

# Menampilkan hasil perhitungan
print("=== PERHITUNGAN DAYA LISTRIK ===")
print(f"Tegangan (V): {tegangan} Volt")
print(f"Arus (I)    : {arus} Ampere")
print(f"Daya (P)    : {daya} Watt")
print(f"Rumus       : P = V × I = {tegangan} × {arus} = {daya} Watt")
```

```python
def hitung_efisiensi_termal(energi_masuk, energi_keluar):
    """
    Menghitung efisiensi termal turbin gas dan menampilkan hasil lengkap.
    """
    efisiensi = (energi_keluar / energi_masuk) * 100

    print("\n=== HASIL PERHITUNGAN EFISIENSI TERMAL ===")
    print(f"Energi Masuk    : {energi_masuk} kJ")
    print(f"Energi Keluar   : {energi_keluar} kJ")
    print(f"Rumus           : ({energi_keluar} / {energi_masuk}) × 100")
    print(f"Efisiensi Termal: {efisiensi:.2f}%")

    return efisiensi

# Contoh pemanggilan
hasil = hitung_efisiensi_termal(1000, 400)
```
