```python
# Program menampilkan nama sebanyak 200 kali
nama = "Nama Anda"  # Ganti dengan nama Anda

for i in range(200):
    print(f"{i+1}. {nama}")
```

```python
# Program menampilkan angka 1 sampai 500
for angka in range(1, 501):
    print(angka)
```

```python
# Menggunakan while loop
angka = 1
while angka <= 500:
    print(angka)
    angka += 1
```

```python
# Program deret bilangan ganjil 1-200 menggunakan while
print("Deret bilangan ganjil dari 1 sampai 200:")

angka = 1
while angka <= 200:
    print(angka, end=" ")
    angka += 2

# Output dalam bentuk vertikal
print("\n\nDeret bilangan ganjil (vertikal):")
angka = 1
while angka <= 200:
    print(angka)
    angka += 2
```

```python
print("=== PROGRAM HITUNG TABUNGAN (FOR LOOP) ===")
total_tabungan = 0

for hari in range(1, 31):
    tabungan_hari_ini = hari * 1000
    total_tabungan += tabungan_hari_ini
    print(f"Hari {hari}: Rp {tabungan_hari_ini:,}")

print(f"\nTotal tabungan setelah 30 hari: Rp {total_tabungan:,}")
```

```python
print("=== PROGRAM HITUNG TABUNGAN (WHILE LOOP) ===")
total_tabungan = 0
hari = 1

while hari <= 30:
    tabungan_hari_ini = hari * 1000
    total_tabungan += tabungan_hari_ini
    print(f"Hari {hari}: Rp {tabungan_hari_ini:,}")
    hari += 1

print(f"\nTotal tabungan setelah 30 hari: Rp {total_tabungan:,}")

# Verifikasi dengan rumus matematika
total_rumus = (30 * 31 // 2) * 1000
print(f"Verifikasi dengan rumus: Rp {total_rumus:,}")
```
